```python
x=df.drop(columns=['Yield_log'])
y=df['Yield_log']
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

X_train.shape,X_test.shape
```

```python
X_train.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit only on training categorical columns
encoder.fit(X_train[['State', 'Crop', 'Season']])

# Transform train and test
X_train_cat = encoder.transform(X_train[['State', 'Crop', 'Season']])
X_test_cat = encoder.transform(X_test[['State', 'Crop', 'Season']])

X_train_cat.shape, X_test_cat.shape
```

```python
# Extract numerical columns
X_train_num = X_train[['Crop_Year', 'Area ']].values
X_test_num = X_test[['Crop_Year', 'Area ']].values

# Combine numerical + categorical
X_train_final = np.hstack((X_train_num, X_train_cat))
X_test_final = np.hstack((X_test_num, X_test_cat))

X_train_final.shape, X_test_final.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

model = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train_final, y_train)

# Predictions
y_pred = model.predict(X_test_final)

# Evaluation

mse = mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

rmse, r2
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

rmse, r2
```

```python
# Train predictions
y_train_pred = model.predict(X_train_final)

train_mse = mean_squared_error(y_train, y_train_pred)
train_rmse = np.sqrt(train_mse)
train_r2 = r2_score(y_train, y_train_pred)

train_rmse, train_r2
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_final, y_train)

# Predictions
y_pred_lr = lr.predict(X_test_final)

mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_test, y_pred_lr)

rmse_lr, r2_lr
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(
    n_estimators=100,
    random_state=42
)

gbr.fit(X_train_final, y_train)

y_pred_gbr = gbr.predict(X_test_final)

mse_gbr = mean_squared_error(y_test, y_pred_gbr)
rmse_gbr = np.sqrt(mse_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

rmse_gbr, r2_gbr
```

```python
df.groupby('Crop')['Yield_log'].mean().sort_values(ascending=False).head(10)
```

```python
# Remove leading/trailing spaces or newlines
df['Crop'] = df['Crop'].str.strip()

# Check Coconut count
df['Crop'].value_counts()['Coconut']
df_no_coconut = df[df['Crop'] != 'Coconut'].copy()

df_no_coconut.shape
```

```python
# Remove leading/trailing spaces from all column names
df_no_coconut.columns = df_no_coconut.columns.str.strip()

# Check columns
df_no_coconut.columns
```

```python
# Separate features and target
X_nc = df_no_coconut.drop(columns=['Yield_log'])
y_nc = df_no_coconut['Yield_log']

# Train/test split
from sklearn.model_selection import train_test_split
X_train_nc, X_test_nc, y_train_nc, y_test_nc = train_test_split(
    X_nc, y_nc, test_size=0.2, random_state=42
)

# Encode categorical features
from sklearn.preprocessing import OneHotEncoder
encoder_nc = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoder_nc.fit(X_train_nc[['State','Crop','Season']])

X_train_cat_nc = encoder_nc.transform(X_train_nc[['State','Crop','Season']])
X_test_cat_nc = encoder_nc.transform(X_test_nc[['State','Crop','Season']])

# Numerical features
import numpy as np
X_train_num_nc = X_train_nc[['Crop_Year','Area']].values
X_test_num_nc = X_test_nc[['Crop_Year','Area']].values

# Combine
X_train_final_nc = np.hstack((X_train_num_nc, X_train_cat_nc))
X_test_final_nc = np.hstack((X_test_num_nc, X_test_cat_nc))

# Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

model_nc = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
model_nc.fit(X_train_final_nc, y_train_nc)

# Predictions
y_pred_nc = model_nc.predict(X_test_final_nc)

# Evaluation
rmse_nc = np.sqrt(mean_squared_error(y_test_nc, y_pred_nc))
r2_nc = r2_score(y_test_nc, y_pred_nc)

rmse_nc, r2_nc
```

```python
feature_names_nc = ['Crop_Year', 'Area'] + list(encoder_nc.get_feature_names_out(['State','Crop','Season']))

importances_nc = model_nc.feature_importances_

importance_df_nc = pd.DataFrame({
    'Feature': feature_names_nc,
    'Importance': importances_nc
}).sort_values(by='Importance', ascending=False)

importance_df_nc.head(10)
```

```python
import matplotlib.pyplot as plt

# Select top 10 features
top10 = importance_df_nc.head(10)

plt.figure(figsize=(10,6))
plt.barh(top10['Feature'][::-1], top10['Importance'][::-1], color='green')
plt.xlabel('Feature Importance')
plt.title('Top 10 Features Affecting Crop Yield (No Coconut)')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(y_test_nc, y_pred_nc, alpha=0.3, color='blue')
plt.plot([y_test_nc.min(), y_test_nc.max()],
         [y_test_nc.min(), y_test_nc.max()],
         'r--', lw=2)
plt.xlabel('Actual Yield_log')
plt.ylabel('Predicted Yield_log')
plt.title('Predicted vs Actual Crop Yield (Random Forest, No Coconut)')
plt.show()
```

```python
# R² values
r2_values = [r2_score(y_train_nc, model_nc.predict(X_train_final_nc)), r2_nc]
labels = ['Train', 'Test']

plt.figure(figsize=(6,5))
plt.bar(labels, r2_values, color=['orange','blue'])
plt.ylim(0,1.05)
plt.ylabel('R²')
plt.title('Train vs Test R² (Random Forest, No Coconut)')
for i, v in enumerate(r2_values):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import joblib

# Save Random Forest model
joblib.dump(model_nc, 'rf_model_no_coconut.pkl')

# Save OneHotEncoder
joblib.dump(encoder_nc, 'encoder_no_coconut.pkl')
```

```python
import streamlit as st
import numpy as np
import joblib

# Load saved model and encoder
model = joblib.load('rf_model_no_coconut.pkl')
encoder = joblib.load('encoder_no_coconut.pkl')

# Streamlit app
st.title("🌾 Crop Yield Prediction")

# Input widgets
state = st.selectbox("Select State", encoder.categories_[0])
crop = st.selectbox("Select Crop", encoder.categories_[1])
season = st.selectbox("Select Season", encoder.categories_[2])
crop_year = st.number_input("Enter Crop Year", min_value=2000, max_value=2030, value=2026)
area = st.number_input("Enter Area (hectares)", min_value=0.01, max_value=1000.0, value=1.0)

# Predict button
if st.button("Predict Yield"):
    # Encode categorical input
    X_cat = encoder.transform([[state, crop, season]])
    X_num = np.array([[crop_year, area]])
    X_input = np.hstack((X_num, X_cat))
    
    # Predict log yield and convert back to actual
    yield_log = model.predict(X_input)[0]
    predicted_yield = np.exp(yield_log) - 1
    
    st.success(f"Predicted Crop Yield: {predicted_yield:.2f} tonnes/hectare")
```
